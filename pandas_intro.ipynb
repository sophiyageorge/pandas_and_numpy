Merge & Concatenate

```python
import pandas as pd
movies = {
    'Movies': ['Ameram',"Daddy cool","Munnarippu",'Immanuel'],
    'Year' : [1991,2009,2014,2013],
    'Director':['Bharathan','Ashiq Abu','Venu Isc','Lal Jose'],
}

index_1 = [1,2,3,4]
df1 = pd.DataFrame(movies,index=index_1)

awards = {
    'Movies' :[ "Ameram",'Mathilukal','Kazhcha','Munnarippu','Varsham','Yathra'],
'Awards':['Film Fare','National','State','Asianet','SIIMA','Special Jury']
}
index_2 = [1,2,3,4,5,6]
df2 = pd.DataFrame(awards,index=index_2)

```

```python
df1
```

```python
df2
```

```python
df1.merge(df2)
```

```python
pd.merge(df1,df2)
```

By default merge is inner join(inter section)

```python
# union merge
df1.merge(df2,how ="right")
# we can also merge based on specific
#  colum if more tha one similar column is present
# df1.merge(df2,on="year")
```

```python
movies = {
    'Movies': ['Ameram',"Daddy cool","Munnarippu",'Immanuel'],
    'Year' : [1991,2009,2014,2013],
    'Director':['Bharathan','Ashiq Abu','Venu Isc','Lal Jose'],
}

index_1 = [1,2,3,4]
df1 = pd.DataFrame(movies,index=index_1)

awards = {
    'Movies' :[ "Ameram",'Mathilukal','Kazhcha','Munnarippu','Varsham','Yathra'],
    'Year':[1992,1991,2005,2015,2005,2008],
'Awards':['Film Fare','National','State','Asianet','SIIMA','Special Jury']
}
index_2 = [1,2,3,4,5,6]
df2 = pd.DataFrame(awards,index=index_2)

```

```python
df1.merge(df2,on='Movies')
```

```python
df1.merge(df2,on='Movies',suffixes=('_released','_awarded'))
```

```python
df1 = pd.DataFrame([['Ammu',10],['Anu',11]],columns=['Name','Age'])
df2 = pd.DataFrame([['Ammu',10,67],['Anu',11,100],['Jini',13,90],['Ani',10,80]],columns = ['Name','Age','Mark'])
df2
```

```python
# by default row wise concatrenation
pd.concat([df1,df2])
```

```python
# colum wise concatnation
result = pd.concat([df1,df2],ignore_index=True)
result
```

```python
df.isnull().sum().sum()
```

Fillna
Fill Null with a constant value

```python
# result = df.fillna(value=0)
result = df.fillna({'Age':df['Age'].min()})
result
```

```python
result.isnull().sum()
```

Fill with previous value

```python
result = df.fillna(method='bfill')
result
```

```python
result = df.fillna(method = 'bfill',axis=1)
result
```
